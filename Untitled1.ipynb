## Definición del Valor Presente Neto (VPN)


El **Valor Presente Neto** es una herramienta financiera fundamental utilizada para evaluar la viabilidad de una inversión. Su propósito es determinar si los flujos de efectivo futuros proyectados, una vez descontados al presente, superan el costo de la inversión inicial.
En términos simples, el VPN permite a los analistas responder a la pregunta: **¿Cuánto valor real agregará este proyecto a la empresa hoy mismo?**
